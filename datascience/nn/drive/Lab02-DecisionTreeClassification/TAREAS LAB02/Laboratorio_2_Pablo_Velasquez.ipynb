```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
import seaborn as sns
sns.set()
```

```python
dataset = pd.read_csv('/Users/Pablo Velasquez/Downloads/Python/irisdata.csv')
sns.pairplot(dataset, hue="class")
X = dataset.drop("class", axis=1)
Y = dataset['class']
```

```python
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X, Y, test_size = 0.33, random_state = 42)
```

```python
from sklearn.tree import DecisionTreeClassifier
classifier = DecisionTreeClassifier(max_depth = 2)
classifier.fit(X, Y)
```

```python
y_pred = classifier.predict(X_test)
y_pred_train = classifier.predict(X_train)
```

```python
clas = DecisionTreeClassifier(max_depth = 3)
y_test = Y.map({'Iris-setosa': 0,'Iris-versicolor': 1, 'Iris-virginica': 2})
clas.fit(X[['petal_width_cm', 'petal_length_cm']], y_test)
plt.figure()
ax = sns.scatterplot(X.petal_width_cm, X.petal_length_cm, hue=Y, palette='Set2')
xlim = ax.get_xlim()
ylim = ax.get_ylim()
absi, orde = np.meshgrid(np.linspace(*xlim, num=200),
                      np.linspace(*ylim, num=200))
Z = clas.predict(np.c_[absi.ravel(), orde.ravel()]).reshape(absi.shape)

desp = ax.contourf(absi, orde, Z, alpha=0.3, cmap = 'Set2')
plt.show()
```
